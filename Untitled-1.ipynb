```python
! pip install edinet_xbrl
! pip install pandas

import glob
from edinet_xbrl.edinet_xbrl_parser import EdinetXbrlParser
import re
import pandas as pd

filenames =["anycolor","monoAI_technology","カバー","グリー株式会社","メルカリ","モイ","株式会社MIXI","株式会社コロプラ","株式会社サイバーエージェント","株式会社ディー・エヌ・エー"]



#中身を抜き出す
"""
パーサー（XBRL読み込みに使うもの）周りを設定
↓
タグの指定を行う

"""
#=====パーサ設定======================================================
#テキストデータ全体のリスト
detail_list = []

for filename in filenames:
    
    parser = EdinetXbrlParser()

    # 企業ごとのテキストデータのリスト
    company_list = []

    xbrl_file = f"xbrl_data/{filename}/XBRL/PublicDoc/*.xbrl"
    xbrl_file = glob.glob(xbrl_file)[0]

    edinet_xbrl_object = parser.parse_file(xbrl_file)
#=====タグ指定=======================================================
    tag_name = ["GovernanceTextBlock","StrategyTextBlock","RiskManagementTextBlock","StrategyAndIndicatorsAndTargetsTextBlock"]

    origin_name = xbrl_file[-60:-30]
    new_name = origin_name.replace("-001","")

    for tag in tag_name:
        tag_list = "{}:{}".format(new_name,tag)



        key = f"{tag_list}"
        context_ref='FilingDateInstant'
        data = edinet_xbrl_object.get_data_by_context_ref(key, context_ref)

        try:
            text_data = data.get_value()
            # 空白（\s）除去
            text_data = re.sub('\s','',text_data)
            # 任意の文字数のタグ（<.*?>）の除去
            text_data = re.sub('<.*?>','',text_data)

            company_list.append(text_data)
            detail_list.append(company_list)
        except AttributeError:
            company_list.append(None)
            detail_list.append(company_list)



# print(detail_list)

columns_list = ['ガバナンス', 'リスク管理', '戦略及び指標と目標', '人的資本に関して' ]
sustainability_df = pd.DataFrame(data=detail_list, columns=columns_list)

display(sustainability_df)

""""
#=====テキストファイルに起こす========================================


            # テキストデータの保存
        text_file = open(f'./sastainability/{filename}_{tag}.txt', 'w', encoding='UTF-8')
        text_file.write(text_data)
        text_file.close()
"""""
```
